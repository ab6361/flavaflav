# Fuction to make the summary file

```python
def bigmod(momentBigdata = [0] * 15,densityBigdata = [0] * 9, density2Bigdata = [0] * 10):
    bigdataF = open('ryan_summary_file.dat', "r")
    string_list = bigdataF.readlines()
    bigdataF.close()
    DaBigData = [0]*31
    i=0
    k=0
    momeStart=0
    densStart=size(momentBigdata)-3
    den2Start=size(densityBigdata)-3+densStart
    othrStart=size(density2Bigdata)-1+den2Start
    while i<size(string_list):
        OriginalFile = string_list[i].split()
        if OriginalFile[0] == momentBigdata[0]:
            j=0
            while j<size(DaBigData):
                if j>momeStart and j<densStart+1:
                    DaBigData[j] = "{0:.3f}".format(momentBigdata[j])
                elif j>othrStart+1 and float(OriginalFile[j]) == 0:
                    DaBigData[j] = momentBigdata[j-othrStart+11]
                else:
                    DaBigData[j] = OriginalFile[j]
                j+=1
            OriginalFile = DaBigData
            k+=1
        if OriginalFile[0] == densityBigdata[0]:
            j=0
            while j<size(DaBigData):
                if j>densStart and j<den2Start+1:
                    DaBigData[j] = "{0:.3f}".format(densityBigdata[j-densStart])
                elif j>othrStart+1 and float(OriginalFile[j]) == 0:
                    DaBigData[j] = densityBigdata[j-othrStart+5]
                else:
                    DaBigData[j] = OriginalFile[j]
                j+=1
            OriginalFile = DaBigData
            k+=1
        if OriginalFile[0] == density2Bigdata[0]:
            j=0
            while j<size(DaBigData):
                if j>den2Start and j<othrStart+1:
                    DaBigData[j] = "{0:.3f}".format(density2Bigdata[j-den2Start])
                else:
                    DaBigData[j] = OriginalFile[j]
                j+=1
            OriginalFile = DaBigData
            k+=1
        if k > 0:
            break
        i+=1
    if k > 0:
        newstring = ' '.join([str(elem) for elem in DaBigData])
        string_list[i] = newstring + '\n'
        bigdataF = open('ryan_summary_file.dat', "w")
        new_file_contents = "".join(string_list)
        bigdataF.write(new_file_contents)
        bigdataF.close()
    if k == 0:
        if momentBigdata[0] == 0 and densityBigdata[0] == 0 and densityBigdata[0] == 0:
            print("no named elemnt")
            return 1
        elif momentBigdata[0] != 0 and densityBigdata[0] != 0 and density2Bigdata[0] != 0 and (momentBigdata[0]!=densityBigdata[0]!=density2Bigdata[0]):
            print("non-matching elemnt name")
            return 1
        else:
            if  momentBigdata[0] != 0:
                DaBigData[0] = momentBigdata[0]
            elif densityBigdata[0] != 0:
                DaBigData[0] = densityBigdata[0]
            elif density2Bigdata[0] != 0:
                DaBigData[0] = density2Bigdata[0]
        j=0
        while j<size(DaBigData):
            if j>momeStart and j<densStart+1:
                DaBigData[j] = "{0:.3f}".format(momentBigdata[j])
            if j>densStart and j<den2Start+1:
                DaBigData[j] = "{0:.3f}".format(densityBigdata[j-densStart])
            if j>den2Start and j<othrStart+1:
                DaBigData[j] = "{0:.3f}".format(density2Bigdata[j-den2Start])
            if j == othrStart+1:
                DaBigData[j] = 0
            if j>othrStart+1:
                if momentBigdata[j-othrStart+11] != 0:
                    DaBigData[j] = momentBigdata[j-othrStart+11]
                else:
                    DaBigData[j] = densityBigdata[j-othrStart+5]
                    
            j+=1
        newstring = ' '.join([str(elem) for elem in DaBigData])
        bigdataF = open('ryan_summary_file.dat', "a")
        bigdataF.write(newstring+'\n')
        bigdataF.close()
```

# calculations and calling bigmod

### momentum

```python
import glob
list_of_files = glob.glob('Momentum_Data/*.momentum.data')
for file_name in list_of_files:
    filename0 = file_name.replace('Momentum_Data\\', '')
    filename1 = filename0.replace('.momentum.data', '')
    MomentaOfKWL = loadtxt(file_name,dtype='str')
    MomentaOfK = np.delete(MomentaOfKWL, 0, 0)
    MomentaOfK = MomentaOfK.astype(float)
    NORM = 4*pi*.1/((2*pi)**3)
    KE = 4*pi*.1/((2*pi)**3)
    MP = MN = 20.65681969394544
    momentBigdata = [0] * 15
    normP = normN = normP1_5 = normP2 = normN1_5 = normN2 = totKeP = totKeN = KeP1_5 = KeP2 = KeN1_5 = KeN2 = numP = numN = 0
    KePsumm = KeP1_5Summ = KeP2Summ = KeNsumm = KeN1_5Summ = KeN2Summ = 0.0
    if size(MomentaOfK[0,:]) == 3 or size(MomentaOfK[0,:]) == 4:
        momentBigdata[0] = filename1
        if size(MomentaOfK[0,:]) == 3:
            yp=MomentaOfK[:,1]
            yn = yp
        elif size(MomentaOfK[0,:]) == 4:
            yp=MomentaOfK[:,3]
            yn = yp
    elif size(MomentaOfK[0,:]) == 5:
        if "N" in MomentaOfKWL[0,1]:
            yn=MomentaOfK[:,1]
            yp=MomentaOfK[:,3]
        if "P" in MomentaOfKWL[0,1]:
            yn=MomentaOfK[:,3]
            yp=MomentaOfK[:,1]
    else:
        print(filename1,"deminsion problem")
        break
    x=MomentaOfK[:,0]
    i=0
    while i<size(x):
        normP += (yp[i]*x[i]**2)*NORM
        KePsumm += (yp[i]*x[i]**4)*KE
        normN += (yn[i]*x[i]**2)*NORM
        KeNsumm += (yn[i]*x[i]**4)*KE
        if x[i] >= 1.5:
            normP1_5 += (yp[i]*x[i]**2)*NORM
            KeP1_5Summ += (yp[i]*x[i]**4)*KE
            normN1_5 += (yn[i]*x[i]**2)*NORM
            KeN1_5Summ += (yn[i]*x[i]**4)*KE
        if x[i] >= 2:
            normP2 += (yp[i]*x[i]**2)*NORM
            KeP2Summ += (yp[i]*x[i]**4)*KE
            normN2 += (yn[i]*x[i]**2)*NORM
            KeN2Summ += (yn[i]*x[i]**4)*KE
        i+=1
    totKeP = MP*KePsumm
    totKeN = MN*KeNsumm
    KeP1_5 = MP*KeP1_5Summ
    KeP2 = MP*KeP2Summ
    KeN1_5 = MN*KeN1_5Summ
    KeN2 = MN*KeN2Summ
    momentBigdata[0] = filename1
    momentBigdata[1] = normP
    momentBigdata[2] = normN
    momentBigdata[3] = normP1_5
    momentBigdata[4] = normP2
    momentBigdata[5] = normN1_5
    momentBigdata[6] = normN2
    momentBigdata[7] = totKeP
    momentBigdata[8] = totKeN
    momentBigdata[9] = KeP1_5
    momentBigdata[10] = KeP2
    momentBigdata[11] = KeN1_5
    momentBigdata[12] = KeN2
    momentBigdata[13] = int(round(normP))
    momentBigdata[14] = int(round(normN))
    bigmod(momentBigdata = momentBigdata)
```

### Density 1-Body

```python
import glob
list_of_files = glob.glob('Density_Data/*.density.data')
for file_name in list_of_files:
    filename0 = file_name.replace('Density_Data\\', '')
    filename1 = filename0.replace('.density.data', '')
    desityOfKWL =loadtxt(file_name,dtype='str')
    desityOfK = np.delete(desityOfKWL, 0, 0)
    desityOfK = desityOfK.astype(float)
    densityBigdata = [0] * 9
    NORM = 4*pi*.1
    normPsumm = densPsumm = rmsPsumm = normNsumm = densNsumm = rmsNsumm = 0
    if size(desityOfK[0,:]) == 3:
        yp = desityOfK[:,1]
        yn = yp
    elif size(desityOfK[0,:]) == 4:
        yp = desityOfK[:,2]
        yn = yp
    elif size(desityOfK[0,:]) == 5:
        if "P" in desityOfKWL[0,1]:
            yp=desityOfK[:,1]
            yn=desityOfK[:,3]
        elif "N" in desityOfKWL[0,1]:
            yn=desityOfK[:,1]
            yp=desityOfK[:,3]
        else:
            print("did not find P or N columns")
            break
    else:
        print("deminsion problem")
        break
    x=desityOfK[:,0]
    i=0
    while i<size(x):
        normPsumm += x[i]**2*(yp[i])*NORM
        #commented out code is density of protons with respect to other protons
        #densPsumm += x[i]**2*(yp[i])**2*NORM
        densPsumm += x[i]**2*(yp[i]+yn[i])*yp[i]*NORM
        rmsPsumm += (x[i]**4)*(yp[i])*NORM
        normNsumm += x[i]**2*(yn[i])*NORM
        #commented out code is density of nutrons with respect to other nutrons
        #densNsumm += x[i]**2*(yn[i])**2*NORM
        densNsumm += x[i]**2*(yp[i]+yn[i])*yn[i]*NORM
        rmsNsumm += (x[i]**4)*(yn[i])*NORM
        i+=1   
    densPsumm = densPsumm/normPsumm
    densNsumm = densNsumm/normNsumm
    rmsPsumm = sqrt(rmsPsumm/normPsumm)
    rmsNsumm = sqrt(rmsNsumm/normNsumm)
    densityBigdata[0] = filename1
    densityBigdata[1] = normPsumm
    densityBigdata[2] = densPsumm
    densityBigdata[3] = rmsPsumm
    densityBigdata[4] = normNsumm
    densityBigdata[5] = densNsumm
    densityBigdata[6] = rmsNsumm
    densityBigdata[7] = int(round(normPsumm))
    densityBigdata[8] = int(round(normNsumm))
    bigmod(densityBigdata = densityBigdata)
```

### Density 2-Body

```python
import glob
list_of_files = glob.glob('Density2_Data/*.density2.data')
for file_name in list_of_files:
    filename0 = file_name.replace('Density2_Data\\', '')
    filename1 = filename0.replace('.density2.data', '')
    desity2OfKWL =loadtxt(file_name,dtype='str')
    desity2OfK = np.delete(desity2OfKWL, 0, 0)
    desity2OfK = desity2OfK.astype(float)
    density2Bigdata = [0] * 10
    NORM = 4*pi*.1
    normPPsumm = normNPsumm = normNNsumm = nearPPsumm = nearNPsumm = nearNNsumm = rmsPPsumm = rmsNPsumm = rmsNNsumm =0
    if size(desity2OfK[0,:]) == 5:
        if desity2OfKWL[0,1] == "RHORPPH":
            ypp = desity2OfK[:,1]
            ynp = desity2OfK[:,3]
            ynn = [0] * size(desity2OfK[:,1])
        elif "RHORNN" in desity2OfKWL[0,1]:
            ypp = [0] * size(desity2OfK[:,1])
            ynp = desity2OfK[:,3]
            ynn = desity2OfK[:,1]
        elif desity2OfKWL[0,1] == "RHORPP" or "RHORPP" in  desity2OfKWL[0,1]:
            ypp = desity2OfK[:,1]
            ynp = desity2OfK[:,3]
            ynn = desity2OfK[:,1]
    elif size(desity2OfK[0,:]) == 4:
        ypp = [0] * size(desity2OfK[:,1])
        ynp = desity2OfK[:,2]
        ynn = [0] * size(desity2OfK[:,1])
    elif size(desity2OfK[0,:]) == 7:
        ypp = desity2OfK[:,1]
        ynp = desity2OfK[:,3]
        ynn = desity2OfK[:,5]
    else:
        print("deminsion problem")
        break
    x = desity2OfK[:,0]
    i=0
    while i<size(x):
        normPPsumm += x[i]**2*ypp[i]*NORM
        normNPsumm += x[i]**2*ynp[i]*NORM
        normNNsumm += x[i]**2*ynn[i]*NORM
        rmsPPsumm += (x[i]**4)*(ypp[i])*NORM
        rmsNPsumm += (x[i]**4)*(ynp[i])*NORM
        rmsNNsumm += (x[i]**4)*(ynn[i])*NORM
        if x[i] < 1:
            nearPPsumm += x[i]**2*ypp[i]*NORM
            nearNPsumm += x[i]**2*ynp[i]*NORM
            nearNNsumm += x[i]**2*ynn[i]*NORM
        i+=1
    if normPPsumm == 0:
            nearPPsumm = 0
            rmsPPsumm = 0
    else:
            nearPPsumm = nearPPsumm/normPPsumm
            rmsPPsumm = sqrt(rmsPPsumm/normPPsumm)
    if normNPsumm == 0:
            nearNPsumm =0
            rmsNPsumm = 0
    else:
            nearNPsumm = nearNPsumm/normNPsumm
            rmsNPsumm = sqrt(rmsNPsumm/normNPsumm)
    if normNNsumm == 0:
            nearNNsumm = 0
            rmsNNsumm = 0
    else:
            nearNNsumm = nearNNsumm/normNNsumm
            rmsNNsumm = sqrt(rmsNNsumm/normNNsumm)
    density2Bigdata[0] = filename1
    density2Bigdata[1] = normPPsumm
    density2Bigdata[2] = normNPsumm
    density2Bigdata[3] = normNNsumm
    density2Bigdata[4] = nearPPsumm
    density2Bigdata[5] = nearNPsumm
    density2Bigdata[6] = nearNNsumm
    density2Bigdata[7] = rmsPPsumm
    density2Bigdata[8] = rmsNPsumm
    density2Bigdata[9] = rmsNNsumm
    bigmod(density2Bigdata = density2Bigdata)
```
