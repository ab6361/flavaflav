## Formatting The File For Use

```python
import glob
list_of_files = glob.glob('NNmomentumq/*.NNmomentumq')
for file_name in list_of_files:
    MoF = open(file_name, "r")
    string_list = MoF.readlines()
    MoF.close()
    editfilename = file_name.replace('NNmomentumq\\', '')
    newfilename = "NNmomentumq_Data/" + editfilename + ".data"
    newfile = open(newfilename, "w")
    i=0
    count = 0
    while i<size(string_list):
        if "****" in string_list[i]:
            count += 1
            j=0
            while j+i+1<size(string_list):
                if j == 0:
                    newfile.write(string_list[j+i-1])
                if not re.search('[0-9]', string_list[j+i+1]):
                    break
                if re.search('[0-9]', string_list[j+i+1]):
                    newfile.write(string_list[j+i+1])
                j+=1
        
        if count > 0:
                break
        i+=1
    newfile.close()
```

## Ploting

```python
import glob
list_of_files = glob.glob('NNmomentumq_Data/*.NNmomentumq.data')
for file_name in list_of_files:
    filename0 = file_name.replace('NNmomentumq_Data\\', '')
    filename1 = filename0.replace('.NNmomentumq.data', '')
    MomentaOfKWL = loadtxt(file_name,dtype='str')
    MomentaOfK = np.delete(MomentaOfKWL, 0, 0)
    MomentaOfK = MomentaOfK.astype(float)
    
    SummaryFile = loadtxt('ryan_summary_file.dat',dtype='str')
    i=0
    while i < size(SummaryFile[:,0]):
        if filename1 in SummaryFile[i,0]:
            sizeRSF = size(SummaryFile[0,:])
            Z = float(SummaryFile[i,sizeRSF-2])
            N = float(SummaryFile[i,sizeRSF-1])
        i+=1
    
    if MomentaOfKWL[0,0] != "K":
        plot(MomentaOfK[:,0], MomentaOfK[:,1]/(N*Z))
        xlabel("Relative Pair Momentum q")
        ylabel(MomentaOfKWL[0,1] + "/(N*Z)")
        title(filename1 + ' ' + MomentaOfKWL[0,1])
        show()
        
        plot(MomentaOfK[:,0], MomentaOfK[:,3]/(Z*(Z-1)/2))
        xlabel("Relative Pair Momentum q")
        ylabel(MomentaOfKWL[0,3] + "/(Z*(Z-1)/2")
        title(filename1 + ' ' + MomentaOfKWL[0,3])
        show()
```
